### MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
x=["python","c","c++","java"]
y=[85,70,69,80]
plt.xlabel("language",fontsize=15)
plt.ylabel("Number",fontsize=14)
plt.title("popularity",fontsize=16)
c=["y","m","r","g"]
plt.bar(x,y,width=0.4,color=c,align="center",edgecolor="black",linewidth=2,linestyle=":",alpha=0.9)
plt.show()
```

```python
import numpy as np
x=["python","c","c++","java"]
y=[85,70,69,80]
z=[23,12,34,56]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("Number",fontsize=14)
plt.title("1st",fontsize=16)

plt.bar(p,y,color="r",width=width,label="popularity")
plt.bar(p1,z,color="y",width=width,label="popularity2")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
x=["python","c","c++","java"]
y=[85,70,69,80]
plt.xlabel("language",fontsize=15)
plt.ylabel("Number",fontsize=14)
plt.title("popularity",fontsize=16)
c=["y","m","r","g"]
plt.barh(x,y,color=c,align="center",edgecolor="black",linewidth=2,linestyle=":",alpha=0.9)
plt.show()
```

### scatter plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,4,5,6,8,9]
y=[2,5,7,12,9,1,7]
plt.scatter(x,y)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("x-axis lable")
plt.ylabel("y-axis lable")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,4,5,6,8,9]
y=[2,5,7,12,9,1,7]
c=["m","y","r","g","black","b","pink"]
s=[50,40,34,56,78,98,21]
plt.scatter(x,y,color=c,sizes=s,alpha=0.9,marker="*",edgecolor="black",linewidth=2)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("x-axis lable")
plt.ylabel("y-axis lable")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,4,5,6,8,9]
y=[2,5,7,12,9,1,7]
z=[2,4,5,6,7,12,21]
c=[10,12,34,56,78,58,46]
s=[50,40,34,56,78,98,21]
plt.scatter(x,y,c=c,cmap="viridis",s=s)
plt.scatter(x,z,c='r',s=s)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("x-axis lable")
plt.ylabel("y-axis lable")
t=plt.colorbar()
t.set_label("color bar")
plt.show()
```

#### pie chart

```python
import matplotlib.pyplot as plt
x=[12,34,56,79]
y=["c","c++","java","python"]
ex=[0.0,0,0.1,0.0]
c=["y","m","b","r"]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",colors=c,shadow=True,radius=0.8,labeldistance=1.1,startangle=90,textprops={"fontsize":9},counterclock=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[12,34,56,79]
y=["c","c++","java","python"]
ex=[0.0,0,0.1,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,radius=0.8,textprops={"fontsize":9},counterclock=False,wedgeprops={"linewidth":5,"edgecolor":"m"}
       ,rotatelabels=True)
plt.title("level graph")
plt.legend(loc=1)

plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[12,34,56,79]
x1=[20,30,45,65]
y=["c","c++","java","python"]
c=["m","r","y","g"]
plt.pie(x,radius=1.5,labels=y)
plt.pie(x1,radius=1,colors="w")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[12,34,56,79]
x1=[20,30,45,65]
y=["c","c++","java","python"]
c=["m","r","y","g"]
plt.pie(x,radius=1.5,labels=y)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)
plt.show()
```

### stem plot

```python
x=[1,2,3,4,5,6]
y=[6,3,3,4,6,7]
plt.stem(x,y,linefmt=":",markerfmt="r*",bottom=2,basefmt="g",label="python",orientation="horizontal")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[5,10,20,30,40,50]
plt.boxplot(x,vert=True,notch=False,widths=0.1,labels=["python"],patch_artist=True,showmeans=True)
plt.legend()
plt.show()
```
